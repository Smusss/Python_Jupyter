Обновим загрузку чтобы исключить изменения, внесееные в фреймворк ранее:

```python
data = pd.read_csv('kc_house_data.csv')
data
```

#### Краткий анализ стоимости объектов недвижимости.  
Для удобства создадим признак - стоимость в тысячах (price_th)

```python
data['price_th'] = round(data['price'] / 1000, 2)
data['price_th'].describe()
```

Стоимость колеблется от 75 до 7700 тысяч, при этом 50% значений распределены на отреске от 322 до 645 тысяч. Значения более 1100 - выбросы.

#### Поиск факторов, влияющих на стоимость начнем с анализа линейных зависимостей.  
Для этого постоим таблицу корреляции:

```python
data_corr= data.drop(columns = ['date', 'waterfront'])
corr_matrix = data_corr.corr()
corr_matrix = np.round(corr_matrix, 2)
corr_matrix[np.abs(corr_matrix) < 0.2] = 0
corr_matrix
```

Наглядно:

```python
plt.figure(figsize=(10, 8))
sns.set(font_scale=0.7)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Correlation matrix');
```

Отдельно выведем значения по критерию price. Отсортируем их по убыванию:

```python
corr_matrix['price'].sort_values(ascending=False).tail(len(corr_matrix['price'])-2).head(10)
```

#### Наиболее высокий коэффициент корреляции выявлен у критериев:  
- sqft_living(0.70) - квадратные метры жилые, а также sqft_living15 (0.59) аналогичный показатель по 15 соседям, и sqft_above (0.61) - кв. метры дома, которые находятся выше земли;
- grade(0.67) - конструкция здания и дизайн. Значение от 1 до 13, где 1-3 плохая конструкция здания и дизайн, 7 - средний уровень конструкции и дизайна, 11-13 - высокое качество конструкции и дизайна;  
- bathrooms (0.53) - количество ванных комнат (0.5 - туалет без душа);  
- view (0.40) - насколько хороший вид,  значение от 0 до 4;  
- квадратные метры дома, которые находятся ниже земли (sqft_basement), количество спален (bedrooms), широта(lat),  количество этажей (floors) - коэффициент составил порядка 0.3.


Помимо линейных зависимостей необходимо проанализировать критерии, которые могут не иметь прямой линейной зависимости, но также могут оказывать влияние на стоимость объекта. Например год постройки и реновация (yr_built, yr_renovated).  


#### Таким образом, для анализа факторов влияния нам необходимо проанализировать:

1. Влияние факторов, предварительно показавших высокую линейную зависимость:  
1.1 Квадратные метры жилые, а также аналогичный показатель по 15 соседям, и метры дома, которые находятся выше земли.  
1.2. Оценка конструкции здания и дизайна.  
1.3. Количество ванных комнат, количество спален. 
1.4. Оценка вида.  
1.5. Площадь подвала, широта расположения объекта, его этажность.  
2. Влияние факторов прочих факторов:  
2.1. Год постройки и реновации.  


#### По порядку:

#### 1.1 Квадратные метры жилые, а также аналогичный показатель по 15 соседям, и метры дома, которые находятся выше земли.

```python
plt.figure(figsize=(6, 4))
plt.scatter(data['sqft_living'], data['price_th'])
plt.title('График разброса значений')
plt.xlabel('Площадь жилая,кв. м.')
plt.ylabel('Стоимость, в тыс.');
```

```python
sns.jointplot(x=data['price_th'], y=data['sqft_living'], kind='reg');
plt.title('График разброса значений')
plt.xlabel('Стоимость, в тыс.')
plt.ylabel('Площадь жилая,кв. м.');
```

Введем дополнительный показатель стоимость квадратного метра жилой площади:

```python
data['price_mliv'] = data['price_th'] / data['sqft_living']
data['price_mliv'].describe()
```

Сравнительные таблицы:  
Стоимость объекта / Доля относительно максимальной цены объекта / Стоимость метра / Доля относительно максимальной цены метра

```python
data['prc_price_mliv'] = data['price_mliv'] / data['price_mliv'].max()
data['prc_price_th'] = data['price_th'] / data['price_th'].max()
data[['price_th', 'prc_price_th', 'price_mliv', 'prc_price_mliv']]. sort_values(by='price_th', ascending=False). head(10)
```

```python
data[['price_th', 'prc_price_th', 'price_mliv', 'prc_price_mliv']]. sort_values(by='price_mliv', ascending=False). head(10)
```

Согласно сравнительным таблицам, стоимость квадратного метра самых дорогих квартир составляет не более 87% от максимальной стоимости квадратного метра в выборке.  

Рассмотрим на графике квартиры каких площадей имеют самые дорогие квадратные метры:

```python
sns.jointplot(x=data['price_mliv'], y=data['sqft_living'], kind='reg');
plt.title('График разброса значений')
plt.xlabel('Стоимость квадратного метра, в тыс.')
plt.ylabel('Площадь жилая,кв. м.');
```

#### Вывод: сильное влияние на стоимость объекта оказывает его площадь, но важно заметить что у объектов площадью более 2000 кв.метров влияние ослабевает - стоимость за квадратный метр начинает падать

```python
sqft = data[['sqft_living','sqft_lot', 'sqft_above', 'sqft_basement', 'sqft_living15', 'sqft_lot15']]
corr_matrix = sqft.corr()
corr_matrix = np.round(corr_matrix, 2)
corr_matrix[np.abs(corr_matrix) < 0.2] = 0
corr_matrix
```

```python
plt.figure(figsize=(4, 3))
sns.set(font_scale=0.7)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')
plt.title('Correlation matrix');
```
