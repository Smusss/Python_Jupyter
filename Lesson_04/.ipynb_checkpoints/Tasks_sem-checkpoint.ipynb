```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv('laptops_10.csv')
data.head()
```

#### 1.1 Изучите распределение количества памяти (Memory_Amount) с помощью matplotlib

```python
plt.figure(figsize=(9,5))
plt.hist(data['Memory_Amount'])
plt.title('Memory amount')
plt.xlabel('GB')
plt.ylabel('Amount');
```

#### 1.2 Изучите распределение стоимости ноутбуков (Price_euros) с помощью matplotlib

```python
plt.figure(figsize=(9,5))
plt.hist(data['Price_euros'], color='#AA0000')
plt.title('Price')
plt.xlabel('Price, euros')
plt.ylabel('Amount');
```

#### 1.3 Изучите распределение веса ноутбуков (Weight) с помощью matplotlib

```python
plt.figure(figsize=(9,5))
plt.hist(data['Weight'], color='#AA9000')
plt.title('Weight')
plt.xlabel('Weight, kg')
plt.ylabel('Amount');
```

#### 2.1 Изучите распределение типов носителя (Memory_Type)

```python
memory_data = data['Memory_Type'].value_counts()
m_type = memory_data.index
values = memory_data.values

plt.figure(figsize=(5,5))
plt.title('Memory_Type')
#plt.pie(memory_data);
plt.pie(values, autopct='%.1f%%', labels = m_type);
```

#### 2.2 Изучите распределение компаний производителей (Company)

```python
company_data = data['Company'].value_counts()
c_type = company_data.index
values = company_data.values

plt.figure(figsize=(5,5))
plt.bar(c_type, values, color = '#AA8890')
plt.title('Companies')
plt.xlabel('Company')
plt.ylabel('Amount')
plt.xticks(rotation=70);
```

#### 2.3 Изучите распределение операционной системы (OpSys)

```python
opsys_data = data['OpSys'].value_counts()
names = opsys_data.index
values = opsys_data.values

plt.figure(figsize=(5,5))
plt.bar(names, values, color = '#AA9990', edgecolor = 'black')
plt.title('Systems')
plt.xlabel('OpSys')
plt.ylabel('Amount')
plt.xticks(rotation=70);
```

#### 2.4 Изучите распределение компаний производителей CPU (Cpu_Company)

```python
cpu_data = data['Cpu_Company'].value_counts()
names = cpu_data.index
values = cpu_data.values

plt.figure(figsize=(5,5))
plt.bar(names, values, color = '#AA3390', edgecolor = 'black')
plt.title('Cpu_Companies')
plt.xlabel('Cpu_Companies')
plt.ylabel('Amount')
plt.xticks(rotation=70);
```

```python
cpu_data = data['Cpu_Company'].value_counts()
names = cpu_data.index
values = cpu_data.values

fig, ax = plt.subplots()
ax.pie(values, labels = names, autopct='%.1f%%', wedgeprops=dict(width=0.8));
```

#### Изучите взаимосвязь компаний производителей ноутбуков и компаний производителей процессоров, 
используя сложенную или многорядовую столбчатую диаграмму  
Процессоры от Samsung не изучайте

#### 3.1 Постройте график в абсолютных величинах

```python
com_cpu = pd.crosstab(index = data['Company'], columns = data['Cpu_Company'])
com_cpu = com_cpu.drop(columns = ['Samsung']).reset_index()
com_cpu
```

```python
plt.figure(figsize=(9,5))
n_ticks = np.arange(len(com_cpu['AMD']))
offset = 0.2
w = 0.4
plt.bar(n_ticks - offset, com_cpu['AMD'], width=w)
plt.bar(n_ticks + offset, com_cpu['Intel'], width=w)
plt.title('Laptops and CPU - producers')
plt.xlabel('Produce company')
plt.ylabel('Amount')
plt.legend(['AMD', 'Intel'])
plt.xticks(n_ticks, com_cpu['Company'], rotation = 70);
```

#### 3.2 Постройте график в относительных величинах

```python
com_cpu = pd.crosstab(index = data['Company'], columns = data['Cpu_Company'], normalize = 'index')
com_cpu = com_cpu.drop(columns = ['Samsung']).reset_index()


plt.figure(figsize=(9,5))
plt.bar(com_cpu['Company'], com_cpu['Intel'])
plt.bar(com_cpu['Company'], com_cpu['AMD'], bottom = com_cpu['Intel'])

plt.title('Laptops and CPU - producers')
plt.xlabel('Produce company')
plt.ylabel('Amount')
plt.legend(['AMD', 'Intel'])
plt.xticks(n_ticks, com_cpu['Company'], rotation = 70);
```

#### 4.1 Изучите взаимосвязь стоимости ноутбука и компании производителя процессора

```python
plt.figure(figsize=(9,5))
plt.boxplot([data['Price_euros'], data['Cpu_Company']])
```

```python
plt.figure(figsize=(9,5))
sns.boxplot(x = data['Price_euros'], y = data['Cpu_Company'])
plt. title('Cpu company and price')
plt.xlabel('Price euros')
plt.ylabel('Cpu company');
```

#### 4.2 Изучите взаимосвязь стоимости ноутбука и типа носителя памяти

```python
plt.figure(figsize=(9,5))
sns.boxplot(x = data['Price_euros'], y = data['Memory_Type'], whis=3)
plt. title('Memory type and price')
plt.xlabel('Price euros')
plt.ylabel('Memory type');
```

#### 4.3 Изучите взаимосвязь стоимости ноутбука и кол-ва оперативной памяти

```python
plt.figure(figsize=(9,5))
sns.jointplot(x = data['Price_euros'], y = data['Memory_Amount'], kind='reg')
plt. title('Memory_Amount and price')
plt.xlabel('Price euros')
plt.ylabel('Memory_Amount');
```

#### 4.4 Изучите взаимосвязь стоимости ноутбука и компании производителя

```python
plt.figure(figsize=(9,5))
sns.boxplot(x = data['Price_euros'], y = data['Company'])
plt. title('Company and price')
plt.xlabel('Price euros')
plt.ylabel('Company');
```

#### Постройте матрицу корреляций для таблицы

```python
df = pd.read_csv('laptops_10.csv')

corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```
