```python
import pandas as pd
```

```python
data = pd.read_csv('laptop_price.csv')
data.head(5)
```

##### 1.1 Создать новый признак Cpu_Company,
который будет содержать только название фирмы, которая произвела CPU (Intel, AMD и т.д)

```python
data['Cpu_company'] = data['Cpu'].apply(lambda x: x.split()[0])
data.head(5)
```

##### 1.2 Создать новый признак Memory_Amount, 
который будет содержать только количество Gb памяти без указания типа носителя (128, 1024, 256 и т.д.)

```python
def check_gb(x):
    memory = x.split()[0]
    if memory.endswith('GB'):
        result = memory.replace('GB', '')
    else:
        result = float(memory.replace('TB', '')) * 1024
    return int(result)
```

```python
# data['Memory_Amount'] = data['Memory'].apply(lambda x: x.split()[0]).str.replace('GB', '')
data['Memory_Amount'] = data['Memory'].apply(check_gb)
data.head(5)
```

##### 1.3 Создать новый признак Memory_Type, 
который будет содержать только тип носителя (HDD, SDD и т.д.)

```python
data['Memory_Type'] = data['Memory'].apply(lambda x: x.split()[1:])
data['Memory_Type'] = data['Memory'].apply(lambda x: x.replace(' ', '').split('GB')[-1])
data.head(5)
```

##### Удалите признаки Memory и ScreenResolution

```python
data.drop(columns = 'ScreenResolution', inplace = True)
data.head(5)
```

##### 2.1 Создайте признак SSD, который изначально равен 0

```python
data['SSD'] = 0
data.head(5)
```

##### 2.2 Поставьте в признаке SSD 1, если ноутбук действительно с типом носителя SSD

```python
data.loc[data['Memory_Type'] == 'SSD', 'SSD'] = 1
data.head(5)
```

##### 2.3 Уберите в признаке Weight значения 'kg' и поменяйте его тип данных на вещественный

```python
data['Weight'] = data['Weight'].str.replace('kg', '')
data['Weight'] = data['Weight'].astype('float')

data.head(5)
```

```python
clients = pd.DataFrame({
 'client_id': [45, 32, 67, 33, 43],
 'laptop_id': [506, 398, 710, 120, 1999]
})
```

##### 3.1 Присоедините к таблице clients данные по ноутбукам через метод join

```python
#id_data = data.set_index('laptop_ID')
id_clients = clients.set_index('laptop_id')
clients_j = id_clients.join(data)
clients_j
```

##### 3.2 Присоедините к таблице clients данные по ноутбукам через метод merge

```python
clients_m = clients.merge(data, left_on = 'laptop_id', right_on = 'laptop_ID')
clients_m
```

##### 4.1 Найдите среднюю стоимость ноутбуков в зависимости от компании производителя
Отсортируйте от меньшей стоимости к большей

```python
price_company = data.groupby('Company').agg({'Price_euros': 'mean'}).sort_values('Price_euros', ascending=True)
price_company
```

##### .2 Найдите минимальную, среднюю и максимальную стоимости ноутбуков в зависимости от производителя процессора

```python
price_CPUcompany = data.groupby('Cpu_company').agg({'Price_euros': ['min','mean', 'max']})
price_CPUcompany
```

##### 4.3 Постройте таблицу с подсчетом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ

```python
data.value_counts('Cpu_company', 'Ram')
```

```python
data.pivot_table(index = 'Cpu_company', columns = 'Ram', values = 'Weight', aggfunc = 'count', fill_value = 0)
```

##### 4.4 Постройте таблицу с подсчетом средней стоимости ноутбуков в данных в зависимости от операционной системы и GB памяти

```python
price_OpSys = data.groupby(['OpSys', 'Memory_Amount']).agg({'Price_euros': 'mean'})
price_OpSys
```

```python
price_OpSys = data.groupby(['OpSys', 'Memory_Amount']).agg({'Price_euros': 'mean'}).unstack('Memory_Amount').fillna(0)
price_OpSys
```
